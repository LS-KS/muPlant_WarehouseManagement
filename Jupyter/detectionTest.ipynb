```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

## Import the test image


```python
img = cv2.imread('pallet_3.png')
plt.imshow(img)
plt.show()
```

```python
arucodict = cv2.aruco.getPredefinedDictionary(cv2.aruco.DICT_6X6_250)
corners, ids, rejected = cv2.aruco.detectMarkers(img, arucodict)
print(f"ids: {ids}")
print(f"corners: {corners}")
print(f"rejected: {rejected}")

img = cv2.aruco.drawDetectedMarkers(img, corners, ids)
plt.imshow(img)
plt.show()
```

```python
ims = []
corners = []
ids = []
for i in range(18):
    filename = f"cups/cup{i+1}.png"
    ims.append(cv2.imread(filename))
    cor, id, rej = cv2.aruco.detectMarkers(ims[i], arucodict)
    corners.append(cor)
    ids.append(id)

fig, ax = plt.subplots(6, 3, figsize=(15, 15))
for i in range(18):
    ims[i] = cv2.aruco.drawDetectedMarkers(ims[i], corners[i], ids[i])
    ax[i//3, i%3].imshow(ims[i])
    ax[i//3, i%3].set_title(f"ids: {ids[i]}")
    plt.show()
```
